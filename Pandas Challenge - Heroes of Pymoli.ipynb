```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
csv_path = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(csv_path)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#purchase_data.count()
total_players = len(purchase_data['SN'].value_counts())
print(total_players)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
purchase_data['Item Name'].value_counts()
```

```python
#Calculate number of unique items
unique_items = len(purchase_data['Item Name'].unique())
#print(unique_items)

#find total revenue
total_revenue = purchase_data['Price'].sum()
#print(total_revenue)

#find average price
average_price = purchase_data['Price'].mean()
#print(average_price)

#find total number of purchases
total_purchases = purchase_data['Purchase ID'].count()
#print(total_purchases)

#Create summary dataframe
purchasing_analysis_df = pd.DataFrame({'Total Purchases': total_purchases, 
                                       'Number of Unique Items': unique_items, 
                                       'Average Item Price': average_price, 
                                       'Total Revenue': total_revenue}, index=[0]) 
purchasing_analysis_df.index.name = None 
purchasing_analysis_df.style.format({'Average Item Price':'${:.2f}',
                                    'Total Revenue':'${:.2f}'})
purchasing_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
print(purchase_data['Age'].max())
print(purchase_data['Age'].min())
```

```python
#Create bins based on age
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 49.9]
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

# Slice the data and place it into bins
#x = pd.cut(purchase_data['Age'], bins, labels=group_names).head()

# Place the data series into a new column inside of the DataFrame
purchase_data['Age Groups'] = pd.cut(purchase_data['Age'], bins, labels=group_names)
#print(purchase_data.head())

# Create a GroupBy object
age_group = purchase_data.groupby('Age Groups')
age_group.head()

#Calculate total players by age
age_group_totals = age_group['Age'].count()
#age_group_totals

#Calculate percentage of each age group
age_percentage = (age_group_totals/total_players)*100
#age_percentage

#Create summary data frame
age_demographics_df = pd.DataFrame({'Total Players by Age':age_group_totals,
                                   'Percentage of Players':age_percentage})
age_demographics_df.index.name = 'Age'
age_demographics_df.style.format({'Percentage of Players':'{:.2f}%'})
age_demographics_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Calculate Purchase Count
purchase_per_age = age_group['Purchase ID'].count()
purchase_per_age

#Calculate total purchase value per group
total_value_by_age = age_group['Price'].sum()
#total_value_by_age

avg_price_per_age = age_group['Price'].mean()

#Calculate Avg Purchase Price per age group
avg_price_per_person_age = total_value_by_age/purchase_per_age
avg_price_per_person_age

#calculate avg total purchase price per person per group
age_purchasing_df = pd.DataFrame({'Total Purchases':purchase_per_age,
                                   'Total Spent':total_value_by_age,
                                'Average Purchase Price':avg_price_per_age,
                                 'Average Purchase Price per Person':avg_price_per_person_age})
age_purchasing_df.index.name = 'Age'
age_purchasing_df.style.format({'Total Spent':'${:.2f}',
                                 'Average Purchase Price':'${:.2f}',
                               'Average Purchase Price per Person': '${:.2f}'})
age_purchasing_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
sn_analysis_df = purchase_data.groupby('SN')
#sn_analysis_df.head()

#Calculate purchases per SN
purchases_per_sn = sn_analysis_df['Purchase ID'].count()
#purchases_per_sn

#Calculate total purchase price per SN
total_spent_per_sn = sn_analysis_df['Price'].sum()
#total_spent_per_sn

#Calculate average price per SN
average_spent_per_sn = sn_analysis_df['Price'].mean()
average_spent_per_sn

top_spenders_df = pd.DataFrame({'Purchase Count':purchases_per_sn,
                               'Average Purchase Price':average_spent_per_sn,
                               'Total Purchase Value':total_spent_per_sn})
#top_spenders_df.index.name = 'SN'
formatted_top_spenders = top_spenders_df.sort_values(['Total Purchase Value'], ascending=False)
formatted_top_spenders.style.format({'Average Purchase Price':'${:.2f}',
                               'Total Purchase Value':'${:.2f}'})
formatted_top_spenders
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data_items = purchase_data[['Purchase ID', 'Item Name', 'Price']]
#purchase_data_items.head()

item_analysis = purchase_data_items.groupby(['Purchase ID','Item Name'])

#Calculate purchase count
item_count = item_analysis['Purchase ID'].count()
#item_count

#Calculate purchase value
item_value = item_analysis['Price'].sum()
#item_value

#Find price of item
price_of_item = item_value/item_count
#price_of_item

#Create summary table
popular_items_df = pd.DataFrame({'Purchase Count':item_count,
                                 'Item Price':price_of_item,
                               'Total Purchase Value': item_value})
popular_items_formatted = popular_items_df.sort_values(['Purchase Count'], ascending=False)
popular_items_formatted.style.format({'Item Price':'${:.2f}',
                                 'Total Purchase Value':'${:.2f}'})
popular_items_formatted                                
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
popular_items_formatted = popular_items_df.sort_values(["Total Purchase Value"],
                                                   ascending=False)
popular_items_formatted.style.format({'Item Price':'${:.2f}',
                                 'Total Purchase Value':'${:.2f}'})
popular_items_formatted                                
```
